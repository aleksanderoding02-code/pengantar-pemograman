### (Nilai Eigen dan Vektor Eigen)
#### pengertian 
Nilai eigen (eigenvalue) atau nilai karakteristik dan vektor eigen (eigenvector) atau vektor karakteristik dari suatu matriks 𝐴 berukuran 𝑁×𝑁 didefinisikan sebagai nilai skalar 𝜆 dan vektor tak nol 𝑣 yang memenuhi persamaan:

$$ Av = \lambda v $$
atau dapat juga ditulis sebagai :
$$ (A - \lambda I)v = 0$$
keterangan :

𝐴 = matriks berukuran 𝑁×𝑁

𝑣= vektor eigen (vektor tak nol)

𝜆 = nilai eigen

𝐼= matriks identitas

0 = vektor nol

Artinya:

Ketika matriks 
𝐴 dikalikan dengan vektor, hasilnya sama dengan vektor 𝑣 yang dikalikan dengan skalar 𝜆. Dengan kata lain, arah vektor tidak berubah, hanya panjangnya yang berubah.


#### Contoh perhitungan
Misalkan matriks

$$
A =
\begin{bmatrix}
0 & 2 \\
1 & 1
\end{bmatrix}
$$

#### Langkah 1: Membentuk matriks λ I

$$
A-\lambda I =
\begin{bmatrix}
0-\lambda & 2 \\
1 & 1-\lambda
\end{bmatrix}
$$
#### Langkah 2: Menghitung determinan

$$
|A-\lambda I| =
\begin{vmatrix}
-\lambda & 2 \\
1 & 1-\lambda
\end{vmatrix}
$$

$$
= (-\lambda)(1-\lambda) - (2)(1)
$$

$$
= \lambda^2 - \lambda - 2
$$


#### Langkah 3: Menentukan Eigenvalue

$$
\lambda^2 - \lambda - 2 = 0
$$

Faktorkan:

$$
(\lambda +1)(\lambda -2)=0
$$

Sehingga diperoleh:

$$
\lambda_1 = -1
$$

$$
\lambda_2 = 2
$$


#### 4. Menentukan Eigenvector

Setelah mendapatkan eigenvalue, kita masukkan ke persamaan:

$$
(A-\lambda I)v = 0
$$


#### Eigenvector untuk λ = -1 

Substitusi ke persamaan:

$$
(A + I)v = 0
$$

Hasil eigenvector:

$$
v_1 =
\begin{bmatrix}
-2 \\
1
\end{bmatrix}
$$

Jika dinormalisasi:

$$
v_1 =
\frac{1}{\sqrt{5}}
\begin{bmatrix}
-2 \\
1
\end{bmatrix}
$$


#### Eigenvector untuk λ = 2 

Substitusi:

$$
(A-2I)v = 0
$$

Hasilnya:

$$
v_2 =
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$
Jika dinormalisasi:

$$
v_2 =
\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$


#### Contoh Soal Eigenvalues dan Eigenvectors


#### Contoh Soal 1

Diberikan matriks

$$
A =
\begin{bmatrix}
4 & 1 \\
2 & 3
\end{bmatrix}
$$

Tentukan:

1. Persamaan karakteristik matriks \(A\)  
2. Nilai eigen (eigenvalues)  
3. Vektor eigen (eigenvectors)


#### Penyelesaian

### 1. Membentuk persamaan karakteristik

Gunakan persamaan:

$$
\det(A - \lambda I) = 0
$$

Maka

$$
A - \lambda I =
\begin{bmatrix}
4-\lambda & 1 \\
2 & 3-\lambda
\end{bmatrix}
$$

Hitung determinannya:

$$
\begin{vmatrix}
4-\lambda & 1 \\
2 & 3-\lambda
\end{vmatrix}
= (4-\lambda)(3-\lambda) - (2)(1)
$$

$$
= 12 -4\lambda -3\lambda + \lambda^2 -2
$$

$$
= \lambda^2 -7\lambda +10
$$

Sehingga persamaan karakteristiknya

$$
\lambda^2 -7\lambda +10 = 0
$$

### 2. Menentukan Eigenvalue

Faktorkan:

$$
(\lambda-5)(\lambda-2)=0
$$

Sehingga

$$
\lambda_1 = 5
$$

$$
\lambda_2 = 2
$$

### 3. Menentukan Eigenvector

Gunakan persamaan

$$
(A-\lambda I)v = 0
$$


#### Eigenvector untuk λ= 5 

$$
A-5I =
\begin{bmatrix}
-1 & 1 \\
2 & -2
\end{bmatrix}
$$

Misalkan

$$
v =
\begin{bmatrix}
x \\
y
\end{bmatrix}
$$

Maka

$$
- x + y = 0
$$

Sehingga

$$
y = x
$$

Eigenvectornya

$$
v_1 =
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$


#### Eigenvector untuk λ = 2 

$$
A-2I =
\begin{bmatrix}
2 & 1 \\
2 & 1
\end{bmatrix}
$$

Maka

$$
2x + y = 0
$$

Sehingga

$$
y = -2x
$$

Eigenvectornya

$$
v_2 =
\begin{bmatrix}
1 \\
-2
\end{bmatrix}
$$


#### Kesimpulan Soal 1

Eigenvalue matriks \(A\) adalah

$$
\lambda_1 = 5 , \quad \lambda_2 = 2
$$

Eigenvectornya

$$
v_1 =
\begin{bmatrix}
1 \\
1
\end{bmatrix}
,
\quad
v_2 =
\begin{bmatrix}
1 \\
-2
\end{bmatrix}
$$


#### Contoh Soal 2

Diberikan matriks

$$
B =
\begin{bmatrix}
2 & 0 \\
0 & 3
\end{bmatrix}
$$

Tentukan eigenvalue dan eigenvector dari matriks tersebut.


#### Penyelesaian

#### 1. Persamaan karakteristik

Gunakan

$$
\det(B-\lambda I)=0
$$

Maka

$$
B-\lambda I =
\begin{bmatrix}
2-\lambda & 0 \\
0 & 3-\lambda
\end{bmatrix}
$$

Determinannya

$$
(2-\lambda)(3-\lambda)=0
$$


### 2. Menentukan Eigenvalue

Maka

$$
\lambda_1 = 2
$$

$$
\lambda_2 = 3
$$


### 3. Menentukan Eigenvector

Gunakan

$$
(B-\lambda I)v=0
$$

#### Eigenvector untuk λ = 2 

$$
B-2I =
\begin{bmatrix}
0 & 0 \\
0 & 1
\end{bmatrix}
$$

Maka

$$
y = 0
$$

Eigenvectornya

$$
v_1 =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

#### Eigenvector untuk λ = 3 

$$
B-3I =
\begin{bmatrix}
-1 & 0 \\
0 & 0
\end{bmatrix}
$$

Maka

$$
x = 0
$$

Eigenvectornya

$$
v_2 =
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$


#### Kesimpulan Soal 2

Eigenvalue matriks \(B\)

$$
\lambda_1 = 2
$$

$$
\lambda_2 = 3
$$

Eigenvector

$$
v_1 =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
,
\quad
v_2 =
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

#### Intuisi Konsep

Eigenvector adalah **vektor yang arah transformasinya tidak berubah oleh matriks**, sedangkan eigenvalue adalah **faktor skala perubahan panjang vektor tersebut**.

Secara matematis dituliskan sebagai

$$
Av = \lambda v
$$

#### kesimpulan

Eigenvalues dan eigenvectors membantu kita memahami bagaimana suatu matriks mentransformasikan ruang.
Eigenvector menunjukkan arah khusus yang tidak berubah, sedangkan eigenvalue menunjukkan besar perubahan pada arah tersebut.
