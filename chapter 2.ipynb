```python
def my_sum(a, b):  # Definisi Fungsi: Membuat blok kode yang dapat digunakan kembali dengan parameter 'a' dan 'b'.
    c = a + b      # Operasi Logika: Menjumlahkan nilai parameter dan menyimpannya dalam variabel lokal 'c'.
    return c       # Mengembalikan Nilai: Mengirimkan hasil perhitungan keluar dari fungsi agar bisa digunakan di bagian lain.

res = my_sum(a=2, b=3) # Pemanggilan: Menjalankan fungsi dengan argumen spesifik dan menyimpan hasilnya ke variabel 'res'.
print('the result is ' + str(res)) # Output & Konversi: Mengubah angka menjadi teks (str) agar bisa digabungkan dan ditampilkan.
```

```python
import numpy as np # Akses Library: Memuat fungsi matematika tingkat tinggi dengan nama panggilan 'np'.

# relevant constants
GREEK_P = np.pi # Konstanta: Mengambil nilai Pi (3.14...) yang presisi untuk perhitungan lingkaran/sudut.
EULER_NUMBER = np.e # Konstanta: Mengambil bilangan Euler (2.71...) untuk pertumbuhan eksponensial.

# print greek_p and euler_number on the screen
print("Archimedes' constant is " + str(GREEK_P)) # Output: Menampilkan teks dan nilai Pi yang sudah diubah ke string.
print("Euler's number is " + str(EULER_NUMBER)) # Output: Menampilkan teks dan nilai e yang sudah diubah ke string.

# trigonometric functions
x = np.sin(GREEK_P / 2) # Kalkulasi: Menghitung nilai sinus dari sudut dalam satuan radian.

# print the result on the screen
print("The sine of a quarter of radiant is " + str(x)) # Output: Menampilkan hasil akhir perhitungan trigonometri.


# defining a 1D array in numpy
my_array = np.array([4, 8, 27]) # Struktur Data: Membuat 'Array' (wadah angka) yang jauh lebih cepat dari list biasa.
# print myArray on the screen
print("myArray is equal to " + str(my_array)) # Output: Menampilkan seluruh isi kumpulan angka dalam array.

log10_my_array = np.log10(my_array) # Vektorisasi: Menghitung logaritma basis 10 untuk SETIAP angka di dalam array secara otomatis.

# print the result on the screen
print("The base-10 logarithms of the elements in myArray are")
print(log10_my_array) # Output: Menampilkan array baru yang berisi hasil logaritma masing-masing elemen.
```
