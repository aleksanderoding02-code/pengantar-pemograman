```python
import numpy as np  # library numerik untuk operasi matematika

def integrate_rec(f, a, b, n):
    # fungsi untuk menghitung integral menggunakan metode rectangle
    h = (b-a)/n  # lebar tiap interval (delta x)
    x = np.linspace(a, b, n+1)  # membagi interval menjadi n bagian
    i = 0  # inisialisasi indeks
    area = 0  # inisialisasi luas
    
    while i < n:  # perulangan untuk menjumlahkan luas tiap rectangle
        sup_rect = f(x[i]) * h  # luas rectangle (tinggi × lebar)
        area += sup_rect  # menjumlahkan ke total luas
        i += 1  # pindah ke interval berikutnya
        
    return area  # mengembalikan hasil integral

'''
Pengujian metode rectangle menggunakan fungsi sin(x)
Integral tentu dari 0 sampai π/2 bernilai 1
'''

sup_5 = integrate_rec(np.sin, 0, np.pi/2, 5)  # n=5 (akurasi rendah)
sup_10 = integrate_rec(np.sin, 0, np.pi/2, 10)  # n=10 (lebih baik)
sup_100 = integrate_rec(np.sin, 0, np.pi/2, 100)  # n=100 (lebih akurat)

print('Menggunakan n=5, hasil metode rectangle = {:.2f}'.format(sup_5))
print('Menggunakan n=10, hasil metode rectangle = {:.2f}'.format(sup_10))
print('Menggunakan n=100, hasil metode rectangle = {:.2f}'.format(sup_100))
```

```python
import numpy as np  # library numerik

def integrate_trap(f, a, b, n):
    # fungsi untuk menghitung integral dengan metode trapezoidal
    h = (b-a)/n  # lebar tiap interval
    x = np.linspace(a, b, n+1)  # membagi interval
    
    i = 1  # mulai dari titik kedua
    area = h*(f(x[0]) + f(x[n]))/2  # menghitung kontribusi ujung (trapesium pertama & terakhir)
    
    while i < n:  # menjumlahkan titik tengah
        sup_rect = f(x[i]) * h  # luas bagian tengah
        area += sup_rect
        i += 1
        
    return area  # hasil integral

'''
Pengujian metode trapezoidal menggunakan fungsi sin(x)
Integral dari 0 sampai π/2 bernilai 1
'''

sup_5 = integrate_trap(np.sin, 0, np.pi/2, 5)  # n=5
sup_10 = integrate_trap(np.sin, 0, np.pi/2, 10)  # n=10

print('Menggunakan n=5, hasil trapezoidal = {:.2f}'.format(sup_5))
print('Menggunakan n=10, hasil trapezoidal = {:.2f}'.format(sup_10))
```

```python
import numpy as np  # Library numerik
from scipy import integrate  # Library integrasi

x = np.linspace(0, 9, 3)  # Membuat titik
y = x**2  # Fungsi

sup_trapz = integrate.trapezoid(y, x)  # Pengganti trapz (versi baru scipy)
sup_simps = integrate.simpson(y, x)  # Simpson versi baru

print('Using n=2, the trapezoidal rule returns a value of {:.0f}'.format(sup_trapz))  # Output trapesium
print('Using n=2, the composite Simpson rule returns a value of {:.0f}'.format(sup_simps))  # Output Simpson

# Output:
# Using n=2, the trapezoidal rule returns a value of 273
# Using n=2, the composite Simpson rule returns a value of 243
```

```python
import numpy as np  # Mengimpor numpy untuk membuat array dan operasi numerik
from scipy import integrate  # Mengimpor modul integrate untuk integrasi numerik

conturs_areas = np.array([194135, 136366, 79745, 38335, 18450, 9635, 3895])  # Data luas penampang (m^2)
x = np.array([0, 25, 50, 75, 100, 125, 150])  # Jarak antar penampang (meter)

vol_traps = integrate.trapezoid(conturs_areas, x)  # Menghitung volume dengan metode Trapesium
vol_simps = integrate.simpson(conturs_areas, x)  # Menghitung volume dengan metode Simpson

print('The trapezoidal rule returns a volume of {:.0f} cubic meters'.format(vol_traps))  # Menampilkan hasil trapesium
print('The composite Simpson rule returns a volume of {:.0f} cubic meters'.format(vol_simps))  # Menampilkan hasil Simpson

# Output:
# The trapezoidal rule returns a volume of 9538650 cubic meters
# The composite Simpson rule returns a volume of 9431367 cubic meters
```

```python
def simple_lithopress(z, ro=2900, g=9.8):  # Fungsi untuk menghitung tekanan litostatik berdasarkan kedalaman
    pressure = z * g * ro / 1e6  # Rumus tekanan: P = z * ρ * g, dibagi 1e6 agar satuan menjadi MPa
    return pressure  # Mengembalikan nilai tekanan

my_pressure = simple_lithopress(z=2000)  # Menghitung tekanan pada kedalaman 2000 meter

print('The pressure at 2000 meters is {0:.0f} MPa'.format(my_pressure))  # Menampilkan hasil tekanan

# Output:
# The pressure at 2000 meters is 57 MPa
```

```python
import numpy as np  # Mengimpor numpy untuk operasi numerik dan array
import matplotlib.pyplot as plt  # Mengimpor matplotlib untuk visualisasi grafik

r = np.linspace(1, 6400, 6400)  # Membuat jarak dari pusat bumi (1–6400 km)

def density():  # Fungsi untuk membuat distribusi densitas bumi berdasarkan lapisan
    ro_inner_core = np.linspace(13100, 12800, 1220)  # Densitas inti dalam
    ro_outer_core = np.linspace(12200, 9900, 2259)  # Densitas inti luar
    ro_lower_mantle = np.linspace(5600, 4400, 2171)  # Densitas mantel bawah
    ro_upper_mantle = np.linspace(4130, 3400, 720)  # Densitas mantel atas
    ro_crust = np.linspace(3100, 2700, 30)  # Densitas kerak bumi
   
    ro_final = np.concatenate((ro_inner_core, ro_outer_core, ro_lower_mantle, ro_upper_mantle, ro_crust))  # Menggabungkan semua lapisan
    
    return ro_final  # Mengembalikan array densitas lengkap

ro = density()  # Memanggil fungsi untuk mendapatkan data densitas

fig, ax = plt.subplots()  # Membuat figure dan axis untuk plot
ax.plot(r, ro, label="Kerapatan (densitas) di bagian dalam Bumi")  # Plot densitas terhadap jarak dari pusat bumi
ax.set_ylabel(r"Density [Kg/m$^3$]")  # Label sumbu Y
ax.set_xlabel("Jarak dari Pusat Bumi (r) [km]")  # Label sumbu X dalam Bahasa Indonesia
ax.legend()  # Menampilkan legenda
ax.grid()  # Menampilkan grid
plt.show()  # Menampilkan grafik
```

```python
def gravity(r):  # Fungsi menghitung percepatan gravitasi
    
    g = np.zeros(len(r))  # Array kosong untuk hasil gravitasi
    Gr = 6.67408e-11  # Konstanta gravitasi
    r = r * 1000  # Konversi km ke meter
    
    for i in range(1, len(r)):  # Loop tiap radius
        
        r1 = r[0:i]  # Radius sampai titik ke-i
        ro1 = ro[0:i]  # Densitas sampai titik ke-i
        r2 = r1[i-1]  # Radius saat ini
    
        y = ro1 * r1**2  # Bagian integral
        y_int = np.trapezoid(y, r1)  # Integrasi metode trapesium (FIX ERROR)
        
        g1 = ((4 * np.pi * Gr) / (r2**2)) * y_int  # Rumus gravitasi
        g[i] = g1  # Simpan hasil
           
    return g  # Return hasil

g = gravity(r)  # Hitung gravitasi

fig, ax = plt.subplots()  # Buat grafik
ax.plot(r, g)  # Plot hasil
ax.grid()  # Grid
ax.set_ylabel(r'g [m/s$^2$]')  # Label Y
ax.set_xlabel("Jarak dari Pusat Bumi (r) [km]")  # Label sumbu X 
ax.annotate('Gravitasi meningkat lalu menurun seiring perubahan kedalaman', xy=(1500, 5.0), xytext=(1500, 3), fontname="times new roman", arrowprops=dict(arrowstyle="->"))
ax.set_title("Distribusi Gravitasi di Dalam Bumi")  # Judul grafik
plt.show()  # Tampilkan grafik
```

```python
def pressure(r, ro, g):  # Fungsi untuk menghitung tekanan berdasarkan kedalaman
    
    p = np.zeros(len(r))  # Membuat array kosong untuk tekanan
    r = r * 1000  # Mengubah satuan dari km ke meter
    
    for i in range(0, len(r)):  # Loop untuk setiap titik
        
        r1 = r[i:len(r)]  # Radius dari titik ke-i sampai permukaan
        ro1 = ro[i:len(r)]  # Densitas dari titik ke-i ke atas
        g1 = g[i:len(r)]  # Gravitasi dari titik ke-i ke atas
        
        y = ro1 * g1  # Bagian dari rumus tekanan (ρ × g)
        p1 = np.trapezoid(y, r1)  # Integrasi menggunakan metode trapesium (FIX ERROR)
        
        p[i] = p1  # Menyimpan hasil tekanan
        
    return p  # Mengembalikan nilai tekanan

p = pressure(r, ro, g) / 1e9  # Mengubah satuan dari Pa ke GPa
z = np.linspace(6400, 1, 6400)  # Kedalaman dari permukaan ke pusat bumi

fig, ax = plt.subplots()  # Membuat grafik
ax.plot(z, p)  # Plot tekanan terhadap kedalaman
ax.grid()  # Menampilkan grid
ax.set_ylabel('P [GPa]')  # Label sumbu Y (tekanan)
ax.set_xlabel('jarak vertikal suatu titik di bawah permukaan Bumi. [km]')  # Label sumbu X (kedalaman)
ax.set_title("Distribusi Tekanan di Dalam Bumi")  # Judul grafik
ax.annotate('Tekanan meningkat seiring bertambahnya kedalaman', 
            xy=(3000, 150), xytext=(2000, 250),
            arrowprops=dict(arrowstyle="->"))
idx_max = np.argmax(p)
ax.plot(z[idx_max], p[idx_max], 'o')  
ax.annotate('Tekanan maksimum', 
            xy=(z[idx_max], p[idx_max]), 
            xytext=(4000, 300),
            arrowprops=dict(arrowstyle="->"))
plt.show()  # Menampilkan grafik
```
