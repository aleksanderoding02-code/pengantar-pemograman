```python
import pandas as pd                              # Mengimpor library pandas untuk membaca dan mengolah data tabel
from scipy.stats.mstats import gmean, hmean      # Mengimpor fungsi rata-rata geometrik dan harmonik
import matplotlib.pyplot as plt                  # Mengimpor library untuk membuat grafik/visualisasi

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # Membaca file Excel dan mengambil sheet 'Supp_traces'

a_mean = my_dataset.Hf.mean()                    # Menghitung rata-rata aritmatika dari kolom Hf
g_mean = gmean(my_dataset['Hf'])                 # Menghitung rata-rata geometrik dari kolom Hf
h_mean = hmean(my_dataset['Hf'])                 # Menghitung rata-rata harmonik dari kolom Hf
 
print('-------')                                 # Menampilkan garis pembatas agar output rapi
print('arithmetic mean')                         # Menampilkan label rata-rata aritmatika
print("{0:.1f} [ppm]".format(a_mean))            # Menampilkan nilai rata-rata aritmatika dengan 1 angka desimal
print('-------')                                 # Garis pembatas

print('geometric mean')                          # Menampilkan label rata-rata geometrik
print("{0:.1f} [ppm]".format(g_mean))            # Menampilkan nilai rata-rata geometrik
print('-------')                                 # Garis pembatas

print('harmonic mean')                           # Menampilkan label rata-rata harmonik
print("{0:.1f} [ppm]".format(h_mean))            # Menampilkan nilai rata-rata harmonik
print('-------')                                 # Garis pembatas

fig, ax = plt.subplots()                         # Membuat area gambar (figure) dan sumbu (axes) untuk plotting

ax.hist(my_dataset.Hf, bins=50, density=False, edgecolor='red', alpha=0.5, label='Measurements Hist') # Membuat histogram dari data Hf dengan 50 bin, menampilkan jumlah data, garis tepi merah, dan transparansi 0.5

ax.axvline(a_mean, color='skyblue', linestyle='--', linewidth=2, label='Arithmetic mean')  # Membuat garis vertikal pada posisi rata-rata aritmatika
 
ax.axvline(g_mean, color='orange', linestyle='--', linewidth=2, label='Geometric mean')  # Membuat garis vertikal pada posisi rata-rata geometrik

ax.axvline(h_mean, color='yellow', linestyle='--', linewidth=2, label='Harmonic mean')  # Membuat garis vertikal pada posisi rata-rata harmonik

ax.set_xlabel('Hf [ppm]')                        # Memberi label pada sumbu X (nilai Hf dalam ppm)
ax.set_ylabel('Frequency')                       # Memberi label pada sumbu Y (jumlah data/frekuensi)
ax.set_title('Histogram Data Hf')                # Memberi judul pada grafik

ax.legend()                                      # Menampilkan legenda untuk membedakan setiap garis dan histogram
plt.show()                                       # Menampilkan grafik ke layar
```

```python
import pandas as pd                              # Library untuk membaca dan mengolah data
import numpy as np                               # Library untuk operasi numerik
import matplotlib.pyplot as plt                  # Library untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # Membaca file Excel dan mengambil data pada sheet 'Supp_traces'

hist, bin_edges = np.histogram(my_dataset['Hf'], bins=50, density=False)  # Membuat histogram secara manual untuk mendapatkan frekuensi dan batas bin

modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1]) / 2  # Menghitung modus (nilai tengah dari bin dengan frekuensi tertinggi)

print('modal value: {0:.0f} [ppm]'.format(modal_value))  # Menampilkan nilai modus dalam satuan ppm

fig, ax = plt.subplots()                         # Membuat area gambar dan sumbu

ax.hist(my_dataset['Hf'], bins=50, density=False, edgecolor='red', label="Measurements Hist", alpha=0.5)  # Membuat histogram dari data Hf dengan 50 bin

ax.axvline(modal_value, color="orange", label="Modal value", linewidth=2)  # Menambahkan garis vertikal untuk menunjukkan posisi modus

ax.set_xlabel('Hf [ppm]')                        # Label sumbu X
ax.set_ylabel('Frequency')                       # Label sumbu Y (lebih tepat daripada "Histogram Data")
ax.set_title('Histogram Data Hf dengan Modus')   # Judul grafik (sudah diperbaiki)

ax.legend()                                      # Menampilkan legenda
plt.show()                                       # Menampilkan grafik
```

```python
import pandas as pd  # library untuk membaca dan mengolah data
import matplotlib.pyplot as plt  # library untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # membaca file Excel

variance =  my_dataset['Th'].var()  # menghitung varians data Thorium (Th)
stddev =  my_dataset['Th'].std()  # menghitung standar deviasi data Th

print ('-------')  # garis pembatas
print ('Varians Th')  # menampilkan judul varians
print("{0:.0f} [ppm^2]".format(variance))  # menampilkan nilai varians
print ('-------')
print ('Standar Deviasi Th')  # menampilkan judul standar deviasi
print("{0:.0f} [ppm]".format(stddev))  # menampilkan nilai standar deviasi
print ('-------')

fig, ax = plt.subplots()  # membuat figure dan axis

ax.hist(my_dataset.Th, bins= 200, density = True, edgecolor='#001f3f', color='#7FDBFF', label='Distribusi Data Th')  # histogram data Th

ax.axvline(my_dataset['Th'].mean()-stddev, color='#B10DC9', label='Rata-rata - 1σ', linewidth=2)  # garis mean - 1 sigma

ax.axvline(my_dataset['Th'].mean()+stddev, color='#2ECC40', label='Rata-rata + 1σ', linewidth=2)  # garis mean + 1 sigma

ax.axvspan(my_dataset['Th'].mean()-stddev, my_dataset['Th'].mean()+stddev, alpha=0.1, color='#FF851B', label='Rata-rata ± 1σ')  # area antara -1σ dan +1σ

ax.set_xlabel('Th [ppm]')  # label sumbu X (konsentrasi Thorium)

ax.set_ylabel('Probability density')  # label sumbu Y

ax.set_title('Distribusi Thorium (Th) dengan Varians dan Standar Deviasi') #menambahkan judul

ax.legend()  # menampilkan legenda

plt.show()  # menampilkan grafik
```

```python
import pandas as pd  # library untuk membaca data
import numpy as np  # library untuk perhitungan numerik
import matplotlib.pyplot as plt  # library untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # membaca file Excel

q1 = np.percentile(my_dataset.Lu, 25, method = 'midpoint')  # menghitung kuartil 1 (25%)
q3 = np.percentile(my_dataset.Lu, 75, method = 'midpoint')  # menghitung kuartil 3 (75%)

iqr = q3 - q1  # menghitung interquartile range (IQR)

print ('-------')  # garis pembatas
print ('Rentang Interkuartil (IQR): {0:.0f} [ppm]'.format(iqr))  # menampilkan nilai IQR
print ('-------')

fig, ax = plt.subplots()  # membuat figure dan axis

ax.hist(my_dataset.Lu, bins=70, density=True, edgecolor='#001f3f', color="#E9089E", label='Distribusi Data Lu')  # histogram data Lu

ax.axvline(q1, color="#5AF40D", label='Kuartil 1 (Q1)', linewidth=2)  # garis Q1

ax.axvline(q3, color="#F80A0A", label='Kuartil 3 (Q3)', linewidth=2)  # garis Q3

ax.axvspan(q1, q3, alpha=0.1, color='#FF851B', label='Rentang Interkuartil (IQR)')  # area antara Q1 dan Q3

ax.set_xlabel('Lu [ppm]')  # label sumbu X

ax.set_ylabel('Probability density')  # label sumbu Y

ax.set_title('Distribusi Konsentrasi Lutetium (Lu) dan Rentang Interkuartil (IQR)')  # judul grafik yang sesuai

ax.legend()  # menampilkan legenda

plt.show()  # menampilkan grafik
```

```python
import pandas as pd  # library untuk membaca data
import numpy as np  # library untuk perhitungan numerik
import matplotlib.pyplot as plt  # library untuk membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # membaca file Excel

a_mean = my_dataset.Ho.mean()  # menghitung rata-rata (mean) data Ho

median = my_dataset.Ho.median()  # menghitung median data Ho

hist, bin_edges = np.histogram(my_dataset['Ho'], bins=20, density=True)  # membuat histogram untuk mencari modus
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2  # menghitung nilai modus dari histogram

fig, ax = plt.subplots()  # membuat figure dan axis

ax.hist(my_dataset.Ho, bins=4, density=True, edgecolor='#001f3f', color='#7FDBFF', label="Distribusi Data Ho")  # histogram data Ho

ax.axvline(modal_value, color='#FF851B', label='Nilai Modus', linewidth=2)  # garis modus

ax.axvline(median, color='#B10DC9',  label='Nilai Median', linewidth=2)  # garis median

ax.axvline(a_mean, color='#2ECC40',  label='Rata-rata (Mean)', linewidth=2)  # garis mean

ax.set_xlabel('Ho [ppm]')  # label sumbu X

ax.set_ylabel('Probability density')  # label sumbu Y

ax.set_title('Distribusi Konsentrasi Holmium (Ho): Mean, Median, dan Modus')  # judul grafik

ax.legend()  # menampilkan legenda

plt.show()  # menampilkan grafik
```

```python
import numpy as np  # library numerik
from scipy.stats import skew  # fungsi untuk menghitung skewness

# menghitung parameter statistik dasar
mean_ho = my_dataset['Ho'].mean()  # rata-rata (mean)
median_ho = my_dataset['Ho'].median()  # median

# mencari modus menggunakan histogram
hist, bin_edges = np.histogram(my_dataset['Ho'], bins=20, density=True)  # membuat histogram
mode_ho = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1]) / 2  # menghitung nilai modus

# menghitung standar deviasi
std_ho = my_dataset['Ho'].std()  # standar deviasi

# menghitung skewness dengan beberapa metode
pearson_1 = (mean_ho - mode_ho) / std_ho  # Pearson skewness pertama
pearson_2 = 3 * (mean_ho - median_ho) / std_ho  # Pearson skewness kedua
fisher_skew = skew(my_dataset['Ho'])  # Fisher-Pearson skewness

# menampilkan hasil
print('-------')
print("Analisis Skewness Data Holmium (Ho)")
print('-------')
print("Mean (Rata-rata)        : {:.2f}".format(mean_ho))
print("Median                 : {:.2f}".format(median_ho))
print("Modus                  : {:.2f}".format(mode_ho))
print("Standar Deviasi        : {:.2f}".format(std_ho))
print('-------')
print("Pearson Skewness 1     : {:.2f}".format(pearson_1))
print("Pearson Skewness 2     : {:.2f}".format(pearson_2))
print("Fisher-Pearson Skewness: {:.2f}".format(fisher_skew))
print('-------')
```

```python
import pandas as pd  # membaca data
import matplotlib.pyplot as plt  # membuat grafik

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # membaca file

fig, ax = plt.subplots()  # membuat figure dan axis

# mengatur tampilan outlier (titik di luar box)
my_flierprops = dict(markerfacecolor='#ffb347', markeredgecolor='#8b0000', marker='D')  # bentuk diamond

# mengatur garis median
my_medianprops = dict(color='#ff0000', linewidth=3)  # merah tebal

# mengatur box
my_boxprops = dict(facecolor='#90ee90', edgecolor='#006400', linewidth=2)  # hijau soft

# membuat boxplot
ax.boxplot(my_dataset.Ho, patch_artist=True, notch=True, widths=0.5,
           flierprops=my_flierprops, medianprops=my_medianprops, boxprops=my_boxprops)

ax.set_ylabel('Ho [ppm]')  # label sumbu Y

ax.set_xticks([1])  # posisi label X
ax.set_xticklabels(['Seluruh Data Ho'])  # label X

ax.set_title('Boxplot Konsentrasi Holmium (Ho)')  # judul grafik

ax.grid(True, linestyle='--', alpha=0.5)  # menambahkan grid

plt.show()  # menampilkan grafik
```
